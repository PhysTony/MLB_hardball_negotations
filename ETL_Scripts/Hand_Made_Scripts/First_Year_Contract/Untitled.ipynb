```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temporadas = ["March", "April", "May", "Jun", "Julie", "August"]
mex = [.05429, .05392, .05364, .05331, .05300, .05272]
eur = [1.060550, 1.062700, 1.064450, 1.066500, 1.068050, 1.071550]
oil = [75.39, 75.6, 75.68, 75.57, 75.33, 75.02]
```

```python
sns.set_style('white')
fig,axes = plt.subplots(figsize = (18,9))

axes.plot(temporadas,
          mex,
          color = (0/255,64/255,196/255,0.85),
          markerfacecolor = (205/255,114/255,220/255,0.6),
          marker = 'o',
          markersize = 15,
          markeredgecolor = 'red',
          markeredgewidth = 3,
          linewidth = 8)
axes.set_title('Mexican peso future' + '\n',
               fontsize = 35,
               fontweight = 'bold',
               color = 'black')
axes.set_xlabel('Month',
                fontsize = 16,
                color = 'black',
                fontweight = 'bold')
axes.set_ylabel("Exchange rate" + "\n", 
                fontsize = 32,
                color = 'black',
                fontweight = 'bold')
axes.tick_params(labelsize = 30,
                 labelcolor = 'black')
axes.tick_params(axis = 'x',
                 labelrotation = 90)
axes.set(xlabel = None)
#axes.set(ylabel = None)
axes.patch.set_facecolor('white')

plt.savefig("mexican.pdf",format = "pdf")
```
