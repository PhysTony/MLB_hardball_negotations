```python
# Create a sample dataframe
sports_data = pd.DataFrame({'Sport': ['MLB', 'MLB', 'NBA', 'NBA', 'NFL', 'NFL', 'NHL', 'NHL'],
                            'Cotnract value': [7.5, 1.5, 15, 2.4, 3.5, 1, 4, 1.5],
                            'Players': ['Agencia libre', 'Todos', 'Agencia libre', 'Todos',
                                        'Agencia libre', 'Todos', 'Agencia libre', 'Todos']})

# Define a dictionary with the colors to use for each category
fa_colors = {'Agencia libre': 'blue', 'Todos': 'red'}
```

```python
import pandas as pd

audience = {'Sport': ['MLB', 'MLB', 'MLB',
                      'NBA', 'NBA', 'NBA',
                      'NFL', 'NFL', 'NFL',
                      'NHL', 'NHL', 'NHL'],
            'Year': ['2019', '2020', '2021',
                     '2019', '2020', '2021',
                     '2019', '2020', '2021',
                     '2019', '2020', '2021'],
            'Audience': [10.5, 7.3, 17.0,
                         1.1, 9.8, 6.0,
                         16.5, 1.0, 9.9,
                         5.9, 15.2, 0.97]}

audience_data = pd.DataFrame(audience)
                             
                             
# Define a dictionary with the colors to use for each category
year_colors = {'2019':'purple', '2020':'navy', '2021':'firebrick'}
```

```python
sns.set_style('white')
fig,axes = plt.subplots(figsize = (13,9))

# Set legend size
plt.rcParams['legend.fontsize'] = 20

# Create a barplot
g_1 = sns.barplot(data = sports_data,
                  x = 'Sport',
                  y = 'Cotnract value',
                  hue = 'Players',
                  palette = fa_colors,
                  edgecolor = "black",
                  linewidth = 4,
                  errwidth = 2,
                  saturation = .9)
g_1.set_ylabel('Valor Contrato[M$]',
               fontsize = 26,
               fontweight = 'bold',
               color = 'black')
g_1.set_xlabel('Deporte',
               fontsize = 19,
               fontweight = 'bold',
               color = 'black')
g_1.tick_params(labelsize = 28,
                labelcolor = 'black')
g_1.set(xlabel = None)

# Add legend
legend = axes.legend(title = 'Jugadores', fontsize = 22, title_fontsize = 26)

plt.savefig("Visualizations/Descriptive_statistics/Thesis/Intro/Pictures/contract_value.pdf",
            format = "pdf")
```
