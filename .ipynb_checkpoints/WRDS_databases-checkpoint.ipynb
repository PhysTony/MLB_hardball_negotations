```python
#Finds GVKEYS for each company in the DataFrames 
C = pd.DataFrame({"gvkey": companies["gvkey"].unique()}).dropna().astype(int)
D = pd.DataFrame({"gvkey": compensations["gvkey"].unique()}).dropna().astype(int)
#GVKEYS in common
G = []
for key1 in C["gvkey"]:
    for key2 in D["gvkey"]:
        if key1 == key2:
            G.append(key1)
#Compannies in common
print("Companies in common: (", len(G), ")")
for company in compensations[compensations["gvkey"].isin(G)]["companyname"].unique():
    print("\t ", company)
```

```python
#Observations for each DataSet in common companies
ceodb = compensations[compensations["gvkey"].isin(G)].drop_duplicates(subset = ["year", "personid"])
compdb = companies[companies["gvkey"].isin(G)].drop_duplicates(subset = ["cyear", "gvkey"])
print("Observations for CEO compensation:  ", ceodb.shape, "\n Observations for companies' performance ", compdb.shape)
```

```python
#DataFrames merge
output_var = ["gvkey", "cyear", "EBIT", "STKCPA"]
compen_var =["gvkey", "year", "companyname", "personid","stateid", "yearfounded", "yearborn", "csub46", "csub2"]
ceodb = ceodb[compen_var]
compdb = compdb[output_var]
data = ceodb.merge(compdb, left_on = ["gvkey", "year"], right_on = ["gvkey", "cyear"], how = "inner")
#Final number of obervations
print(colored("Obervations in total:  ", "red"), colored(data.shape, "red"))
```

## Create Variables For Regression 

```python
#Reset index of Data frame and replace NA
data_ok = data_ok.reset_index(drop = True)
data_ok = data_ok.replace(np.nan, 0)
data_ok["comp"] = data_ok["STKCPA"]+data_ok["csub46"]/1000000
data_ok.head()
```

```python
data_ok = data_ok.sort_values(["personid","year"]).reset_index(drop = True)
```

```python
data_ok.shape
```

```python
#AQUI TENIA UN ERROR -> Tenia que tomar la diferencia en el tiempo 
def sqrt_dif(X):
    S = []
    for i in range(0, len(X)-1):
        d = np.sqrt(X[i+1])-np.sqrt(X[i])
        S.append(d)
    try:
        S.append(d)
    except: 
        S.append(0)
    return S
Y = []
for p in data_ok["personid"].unique():
    X=data_ok[data_ok["personid"]==p]["comp"].values
    S = sqrt_dif(X)
    Y = np.concatenate((Y, S))
data_ok["Y"] = Y
```

```python
#Get Max output by company y_H
Ma = pd.DataFrame({"gvkey":data_ok.groupby(by = "gvkey")["EBIT"].max().index, "maxebit": data_ok.groupby(by = "gvkey")["EBIT"].max().values})
Min = pd.DataFrame({"gvkey":data_ok.groupby(by = "gvkey")["EBIT"].min().index, "minebit": data_ok.groupby(by = "gvkey")["EBIT"].min().values})
data_ok = data_ok.merge(Ma, on = "gvkey", how = "left")
data_ok = data_ok.merge(Min, on = "gvkey", how = "left")
```

```python
#Get indicator I- (0 when yt is closer to yH than it is closer to 0=yL)
I = []
for y,ma,mi in zip(data_ok["EBIT"], data_ok["maxebit"],data_ok["minebit"]):
    if y > (ma+mi)/2:
        I.append(0)
    else: 
        I.append(1)
data_ok["ind"] = I
data_ok.head()
```

```python
#Create independent variable (-1^I)*(yt/sqrt(yH))
X = []
for i,x in zip((-1)^(data_ok["ind"]),(data_ok["EBIT"]/np.sqrt(data_ok["maxebit"]))):
    X.append(i*x)
data_ok["X"] = X
data_ok.head()
```

```python
#Dave Data for regression (Done in R)
data_ok.to_csv("RegData.csv")
```

## Regression Results

```python
Image("regres.png")
```

Recordemos que el estimador para $X$ es justamente el factor $\varepsilon$ del modelo. 

Se realizaron dos regresiones, un modelo lineal como base y un modelo de efectos fijos "within" que explota la estructura de los datos en panel. En el modelo solo se controla por la edad del agente, la antiguedad de la empresa y el estado en el que se encuentra la empresa. No se agregan ms controles por falta de datos. 

Los resultados de MCO muestran un estimador de $\varepsilon$ positivo y estadisticamente significativo. Podemos ver que incluso con este modelo el estimador es de la magnitud esperada, es decir, es pequeo. 

Cuando usamos un modelo de efectos fijos (por ao y por persona) se logra un mejor ajuste y el estimador es estadisticamente significativo al 1\%. El estimador para $\varepsilon$ es de 0.025 por lo que tambin coincide con la magnitud usada en nuestro modelo terico.
