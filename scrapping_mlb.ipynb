````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
````

Crearemos el objeto url para usarlo en el scrapping:

````python
url = 'https://www.spotrac.com/mlb/arizona-diamondbacks/payroll/'
````

Creamos el objeto de la página y verificaremos si el servidor nos dio permiso (Se tiene que obtener ```<Response [200]>```):

````python
page = requests.get(url)
````

````python
page
````

Obtendremos el objeto de html a lxml para hacer mejor el formato y vobtener toda la información de la página:

**Esto nos sirve para buscar las clases o id de las tablas que necesitamos**. Por lo general son elementos con etiqueta ```<tr>```.

````python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
soup
````

### Scrapping General

Crearemos un objeto para cada una de las tablas encontradas con respecto al equipo de _Arizona Diamondbacks_.

Como hay cuatro tablas con la clase _datatable rtable_, tendremos que crear una lista para almacenar cada una de esas tablas para despues tratar entrada por entrada.

````python
# General table
table = soup.find_all("table", class_= "datatable rtable")
table
````

Veamos que, en efecto, hay más de una tabla con la misma clase:

````python
len(table)
````

Ahora creemos una nueva variable por cada entrada respectiva que contendrá esta:

````python
# Main table
main = table[0]
main
````

````python
# 2022 Suspended/Restricted
suspended = table[1]
suspended
````

````python
# 2022 Minor
minor = table[2]
minor
````

````python
# 2022 Singning
sign = table[3]
sign
````

Ahora las demás tablas que tienen alguna clase única:

````python
# 2022 Payroll Totals
playroll = soup.find("table", class_= "datatable captotal rtable rtable-notsticky")
playroll
````

````python
# 2022 Competitive Balance Tax Totals
balance = soup.find("table", class_= "datatable captotal")
balance
````

### Creación de dataframes

Se tendrán que analizar los _headers_ para quitar o modificar nombres que cuasen problemas para la creación del dataframe; por otro lado, se tendrán que elimnar columnas vacías y quitar los _NaN_ de las columnas.

#### Main table:

````python
# Obtain every title of columns with tag <th>
headers_1 = []
for i in main.find_all("th"):
    title = i.text
    headers_1.append(title)
````

````python
headers_1
````

Tenemos que cambiar algunos nombres y para ello podemos imprimir los índices de cada título como guía:

````python
j=0
for i in headers_1:
    print(str(i) + ": " + str(j))
    j+=1
````

````python
headers_1[0] = "Active Players"
headers_1[9] = "Payroll"
headers_1
````

Luego, creemos el dataframe:

````python
main_ds = pd.DataFrame(columns = headers_1)
````

````python
for j in main.find_all("tr")[1:]:
    row_data = j.find_all("td")
    row = [i.text for i in row_data]
    length = len(main_ds)
    main_ds.loc[length] = row
````

````python
main_ds.head()
````

Quitemos las caractérres innecesarios en la columna *Lux. Tax Salary* y *Waiver Options*:

````python
auxi_1 = main_ds["Lux. Tax Salary"].tolist()
auxi_2 = main_ds["Waiver Options"].tolist()
````

````python
# Veamos si se crearon correctamente las listas
print(auxi_1[0:3])
print(auxi_2[0:3])
````

````python
lux_aux = []
lux = []

# Quitemos los \t
for i in auxi_1:
    auxi_11 = i.replace('\t', '')
    lux_aux.append(auxi_11)

# Quitemos los \n
for i in lux_aux:
    auxi_11 = i.replace('\n', '')
    lux.append(auxi_11)
print(lux)
````

````python
lux = pd.DataFrame(lux)
lux.head()
````

````python
waiver_aux = []
wavier = []

# Quitemos los \t
for i in auxi_2:
    auxi_22 = i.replace('\t', '')
    waiver_aux.append(auxi_22)

# Quitemos los \n
for i in waiver_aux:
    auxi_22 = i.replace('\n', '')
    wavier.append(auxi_22)
print(wavier)
````

````python
wavier = pd.DataFrame(wavier)
wavier.head()
````

````python
print(main_ds.columns)
````

````python
main_ds["Waiver Options"] = wavier
main_ds["Lux. Tax Salary"] = lux
````

````python
main_ds.head()
````
